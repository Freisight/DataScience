```python
import pandas as pd
log = pd.read_csv('other/visit_log.csv', sep=';')
```

```python
log.loc[log['source_type'] != 'traffic_source', 'source_type'] = log['traffic_source']
log.loc[(log['traffic_source'] == 'yandex') | (log['traffic_source'] == 'google'), 'source_type'] = 'organic'
log.loc[(log['region'] == 'Russia') & (log['traffic_source'] == 'paid'), 'source_type'] = 'ad'
log.loc[(log['region'] != 'Russia') & (log['traffic_source'] == 'paid'), 'source_type'] = 'other'
log.loc[(log['region'] == 'Russia') & (log['traffic_source'] == 'email'), 'source_type'] = 'ad'
log.loc[(log['region'] != 'Russia') & (log['traffic_source'] == 'email'), 'source_type'] = 'other'
log.head(20)

# а можно ли через функцию apply пропускать информацию из двух столбцов?
```

```python
# Задание 3

# Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 
# 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp
# для данного значения userId.
```

```python
rate = pd.read_csv('other/ml-latest-small/ratings.csv')
# таков путь:
# groupby userId
# rating count > 100
# livetime = timestamp max - timestamp min
# mean livetime (timestamp)

# diff_time = rate.groupby('userId').agg({'timestamp': ['min', 'max']})
diff_time = rate.groupby('userId').agg({'timestamp': ['min', 'max', 'count']})
# сортировка по юзер айди, но выводим по каждому юзеру минимальное время и максимальное время через agg() который позволит
# вычислить дальше разницу

diff_time['dif'] = diff_time[('timestamp', 'max')] - diff_time[('timestamp', 'min')] #новый столбец с временем жизни
# diff_time = diff_time.drop('timestamp', axis=1)  #удаляем лишнее
diff_time = diff_time[diff_time[('timestamp', 'count')] >= 100]
diff_time[( 'dif','')].mean()

# у меня так и не получилось избавиться от мультииндекса,
# после ресет_индекса толком не было изменений, что помогли бы мне разобраться дальше
# поэжтому вывел значения по колхозному методу. Теперь среднее значение?
```

```python
# Задание 4

# Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе Материалы для лекции
# «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).

# Необходимо сформировать две таблицы:
# таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
# аналогичную таблицу по типам выручки с указанием адреса клиента
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd = rzd.set_index('client_id')
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
auto.set_index('client_id', inplace=True)
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air.set_index('client_id', inplace=True)
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base.set_index('client_id', inplace=True)
```

```python
# тут выводим таблицу всех клиентов без адресов
all = rzd.join(auto, how="outer").join(air, how="outer")
all.fillna('0').reset_index() #нули поменял для вида, для рассчетов нулди надо поменять обратно на NaN
```

```python
# тут выводим полную таблицу всех людей с адресами
all_base = client_base.join(rzd, how='outer').join(auto, how='outer').join(air, how='outer').reset_index()
all_base = all_base.fillna('0')
all_base
```
